## First Column (Filling of NAN values )

```python
df["Adults (ages 15+) and children (ages 0-14) newly infected with HIV"].isna().sum()
```

```python
df["Adults (ages 15+) and children (ages 0-14) newly infected with HIV"].describe()
```

```python
df["Adults (ages 15+) and children (ages 0-14) newly infected with HIV"].fillna(3300.000000,inplace=True)
```

## Fourth Columns

```python
df["Arms exports (SIPRI trend indicator values)"].isna().sum()
```

```python
df["Arms exports (SIPRI trend indicator values)"].describe()
```

```python
# The data is missing in various places, As Arms export cannot be zero in any given country, we must fill other countries with minimal values
df["Arms exports (SIPRI trend indicator values)"].fillna(40000,inplace=True)
```

```python
df["Arms exports (SIPRI trend indicator values)"].isna().sum()
```

## Fifth Column

```python
df["Arms imports (SIPRI trend indicator values)"].isna().sum()
```

```python
df["Arms imports (SIPRI trend indicator values)"].describe()
```

```python
# Arms import cannot be zero, thus using minimal values to fill the NAN's 
df["Arms imports (SIPRI trend indicator values)"].fillna(1000000,inplace=True)
```

```python
df["Arms imports (SIPRI trend indicator values)"].isna().sum()
```
